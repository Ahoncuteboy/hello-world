# HW10：進階版剪刀石頭布

把之前的剪刀、石頭、布遊戲加上以下功能：

1. 檢查使用者出拳是否是在選擇範圍內 (1,2,3)
    - 若使用者輸入非法選項，就顯示「請輸入合法的選項！」
2. 計分板功能
    - 在每一次猜拳之後，電腦會顯示目前玩家與電腦各別贏的次數
    - 提示：可以思考如何用 List 紀錄輸、贏、及平手次數的資料
3. 詢問使用者是否再玩一次的功能
    - 使用者可回應 Yes（Y）或是 No（N）
    - 若使用者回答是 Y，就再玩一次
    - 若使用者回答是 N，就顯示「感謝你玩剪刀石頭布！」並結束游戲

```python
from numpy.random import randint

comp_hand_array = randint(1, 4, size=3)

print("歡迎來到剪刀石頭布！")
name = input("請輸入您的名稱：")
choices = {1: '剪刀', 2: '石頭', 3: '布'}
status = 1

while status == 1:

    score_board = {'win': 0, 'lose': 0, 'even': 0}

    for i in range(3):

        user_hand_num = input(("請出拳： (1) 剪刀 (2) 石頭 (3) 布："))

        while True:
            try:
                user_hand_num = int(user_hand_num)
            except:
                pass
            if user_hand_num not in list(choices.keys()):
                print("請輸入合法的選項！")
                user_hand_num = input(("請出拳： (1) 剪刀 (2) 石頭 (3) 布："))
            else:
                break


        user_hand = choices[user_hand_num]
        comp_hand_num = comp_hand_array[i]
        comp_hand = choices[comp_hand_num]

        if user_hand_num - 1 == comp_hand_num or user_hand_num + 2 == comp_hand_num:
            result = '你贏了一把'
            score_board['win'] += 1
        elif user_hand_num == comp_hand_num:
            result = '平手！'
            score_board['even'] += 1
        else:
            result = '你輸了一把！'
            score_board['lose'] += 1


        discription = f'{name} 出了: {user_hand}, 電腦出了: {comp_hand}'
        print('=================================================================')
        print(discription)
        print('=================================================================')
        print(result)
        print('=================================================================')
        board_upper = '|' + '{:^30}'.format(name) + '|' + '{:^30}'.format('電腦') + '|'
        print(board_upper)
        print('=================================================================')
        board_lower = '|' + '{:^30}'.format(score_board['win']) + '|' + '{:^32}'.format(score_board['lose']) + '|'
        print(board_lower)
        print('=================================================================')

    ending = f"{name}贏了: {score_board['win']}把\n電腦贏了: {score_board['lose']}把\n平手: {score_board['even']}把"
    if score_board['win'] > score_board['lose']:
        final = '恭喜你是最終勝利者！'
    elif score_board['win'] == score_board['lose']:
        final = '平手！'
    else:
        final = '很遺憾，你輸了！'

    print('=================================================================')
    print(ending)
    print('=================================================================')
    print(final)

    again = input('想要再玩一次嗎？ (Y) 是 (N) 否：')
    legel_ans = ['Y', 'y', 'N', 'n']
    while True:
        if again not in legel_ans:
            print("請輸入合法的選項！")
            again = input('想要再玩一次嗎？ (Y) 是 (N) 否：')
        else:
            break

    if again == 'N'or again =='n':
        status = 0
    else:
        status = 1

print('感謝你玩剪刀石頭布！')
```
